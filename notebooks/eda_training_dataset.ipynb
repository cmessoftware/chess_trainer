# Análisis Exploratorio de Datos (EDA) - Chess Training Dataset

```python
# Cargar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuración visual
sns.set(style='whitegrid')
```

```python
# Cargar dataset
import os

TRAINING_DATA_PATH = "/app/src/data/training_dataset_20250603_020653.csv"

df = pd.read_csv(TRAINING_DATA_PATH)
df.shape, df.columns.tolist()
```

## Revisión Inicial

```python
df.info()
df.describe(include='all')
```

## Valores Nulos y Únicos

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.nunique().sort_values(ascending=False)
```

## Distribuciones Numéricas

```python
num_cols = [
    'material_balance', 'material_total', 'num_pieces', 'branching_factor',
    'self_mobility', 'opponent_mobility', 'score_diff', 'move_number_global'
]
df[num_cols].hist(bins=30, figsize=(16, 12))
plt.tight_layout()
```

## Boxplots por Fase y Color

```python
for col in ['branching_factor', 'self_mobility', 'opponent_mobility']:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x='phase', y=col, data=df)
    plt.title(f'{col} por Fase')
    plt.show()
```

## Correlación entre Variables Numéricas

```python
plt.figure(figsize=(12, 10))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

## Análisis de Score y Movilidad

```python
sns.scatterplot(data=df, x='branching_factor', y='score_diff', hue='phase')
plt.title('Score Diff vs Branching Factor')
plt.show()
```

## Jugadas por Número Global

```python
df[df['move_number_global'] <= 10].groupby('move_number_global')['score_diff'].mean().plot(marker='o')
plt.title('Promedio de Score Diff en las Primeras Jugadas')
plt.xlabel('Número de Jugada Global')
plt.ylabel('Score Diff Promedio')
plt.grid(True)
plt.show()
```
